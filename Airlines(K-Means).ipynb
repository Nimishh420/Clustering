```python
df = pd.read_excel('Airlines.xlsx', sheet_name='data')
df
```

```python
df2 = df.drop(['ID#'],axis=1)
df2
```

```python
# Normalize numerical data 
from sklearn.preprocessing import normalize
df_norm=pd.DataFrame(normalize(df2),columns=df2.columns)
df_norm
```

```python
# Create Dendrograms
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(12, 8))  
dendograms=sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
plt.figure(figsize=(12, 8))  
dendograms=sch.dendrogram(sch.linkage(df_norm,'average'))
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**Build Cluster algorithm using K=5**

```python
clusters5=KMeans(5,random_state=30).fit(df_norm)
clusters5
```

```python
clusters5.labels_
```

```python
# Assign clusters to the data set
airline = df2.copy()
airline['clusters5id']=clusters5.labels_
airline
```

```python
# Compute the centroids for K=5 clusters with 11 variables
clusters5.cluster_centers_
```

```python
# Group data by Clusters (K=5)
airline.groupby('clusters5id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airline['clusters5id'],airline['Balance'], c=clusters5.labels_) 
```
